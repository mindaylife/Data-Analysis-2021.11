# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

### 1.Line 그래프

```python
np.random.seed(2021)
df = pd.DataFrame(np.random.randn(100, 3),
                index = pd.date_range('1/1/2021', periods=100),
                columns=list('ABC')).cumsum()
df.plot()
plt.title('Pandas의 plot method 사용 예')
plt.xlabel('일자')                  #index = x좌표
plt.ylabel('데이터')                #value = y좌표
plt.show()
```

### 2.Bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar', rot=0)      #rot = 밑에있는 인덱스 각도 조절
plt.title('Setosa 품종의 sepal length')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = "Feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 feature별 평균')
plt.ylim(0,9)           #y좌표의 리미트 제한을 높여줌
plt.show()
```

### 3. Pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객 수')
plt.show()
```

### 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box(figsize=(8,10))
plt.title("Setosa 품종의 각 Feature별 Box plot")
plt.show()
```

```python
iris.boxplot(by='species', figsize=(10,8))
plt.tight_layout()
plt.show()
```
