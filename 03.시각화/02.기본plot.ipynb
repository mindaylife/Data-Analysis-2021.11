# Matplotilb - 기본 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot([0,1,2,3],[0,1,4,9])       
plt.title('라인 플롯')              
plt.grid()                          
plt.show()   
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')       # color, marker, linestyle
plt.title('Red-circle-Line')                        #color: blue, green, red, cyan, magenta, yellow, black(k), white
plt.grid()                                          #linestyle: 직선(-), dashed(--), dash-dot(-.), dotted(:)
plt.show()                                          #marker: circle(o), 삼각형(^), 역삼각형(v), 사각형(s), 별(*), 플러스(+) 
```

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--')
plt.plot([8,4,2,1], 'g^:')       
plt.title('Margenta-square-Dashed, Green-삼각형-점선')                        
plt.grid()                                          
plt.show()  
```

### line style

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ms--', linewidth=3, markersize=10)
plt.plot([8,4,2,1], 'g^:', lw=5, ms =20)       
plt.title('Margenta-square-Dashed, Green-삼각형-점선')                        
plt.grid()                                          
plt.show()  
```

### 그림 범위

```python
plt.plot([0,1,2,3],[0,1,4,9], 'ro-')       
plt.title('Red-circle-Line')       
plt.xlim(-1,5)
plt.ylim(-1,10)                 
plt.grid()                            
plt.show() 
```

### Tick(틱)설정

```python
X = np.linspace(-np.pi, np.pi, 500)
Y = np.cos(X)
plt.title('Tick설정')
plt.plot(X,Y)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.show()
```

### 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한 플롯')
plt.plot(X,Y, label='cosine')
plt.plot(X,S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.show()
```

### X축 Y축 Label

```python
plt.title('x축, Y축 라벨을 표시한 플롯')
plt.plot(X,Y, label='cosine')
plt.plot(X,S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

### 그림의 크기

```python
plt.figure(figsize=(10,6))
plt.title('그림크기지정')
plt.plot(X,Y, label='cosine')
plt.plot(X,S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

### 그림 저장

```python
plt.figure(figsize=(10,6))
plt.title('그림 크기 지정한 그래프를 저장')
plt.plot(X,Y, label='cosine')
plt.plot(X,S, label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.savefig('삼각함수.png')
```
