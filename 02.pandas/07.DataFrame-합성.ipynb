```python
### 서울시 인구 데이터와 cctv데이터 합치기
pop = pd.read_csv("서울시 자치구별 주민등록 인구.txt", sep='\t', skiprows=2)
pop
```

```python
# 필요한 컬럼만 가져오기
pop = pop[['자치구','계','65세이상고령자']]
pop.head(3)
```

```python
# 합계 지우기
pop.drop([0], inplace=True)
pop.head(3)
```

```python
# 자치구를 인덱스로, 컬럼명 변경
pop.set_index('자치구', inplace=True)
pop.rename(columns={'계': '총인구', '65세이상고령자':'노인인구'}, inplace=True)
pop.head(3)
```

```python
# 숫자를 정수로 변환
pop = pop.apply(lambda x: x.str.replace(',','')).astype(int)
pop.head(3)
```

```python
# 노인인구 비율
pop['노인인구 비율'] = (pop.노인인구 / pop.총인구 * 100).round(2)
pop.head(3)
```

```python
# cctv 파일 읽기
cctv = pd.read_csv('cctv.csv')
cctv.head(3)
```

```python
# index
cctv.set_index('자치구', inplace=True)
cctv.head(3)
```

```python
df = pop.join(cctv)
df.head()
```

```python
# 인구 1000명당 cctv 댓수
df['1천명당 cctv댓수'] = (df.CCTV댓수 / df.총인구 *1000).round(2)
df.tail()
```

```python
df.to_csv('서울시 자치구별 인구와 CCTV.csv')
```
